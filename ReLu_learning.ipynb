```python
# Rectifier Linear (ReLu)

import numpy as np
import matplotlib.pyplot as plt

f=np.linspace(-10,10,50)
def ReLu(f):
  return np.maximum(f,0)
g=ReLu(f)
plt.plot(f,g)
plt.xlabel("f")
plt.ylabel("g")
```
