```python
# sigmoid
import numpy as np
import matplotlib.pyplot as plt
f=np.linspace(-10,10,50)
def sigmoid(f):
  return (1/(1+np.exp(-f)))
g=np.round(sigmoid(f),3)
print(g)

plt.plot(f,g)
plt.xlabel('f')
plt.ylabel("g(sigmoid)")

```
